```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv("Raw Sales Data.csv")

print(df.info())

df.head()
```

```python
#1
df ['Qty'] = df ['Qty'].str. replace (',','')
df['Qty'] = df['Qty'].astype(int)
df1=pd.DataFrame(df.groupby(["Month"])["Qty"].sum())
df1
```

```python
#2
df2=pd.DataFrame(df.groupby(["Category"])["Qty"].sum())
df2
```

```python
#3
df ['Sales'] = df ['Sales'].str. replace (',','')
df['Sales'] = df['Sales'].astype(int)
df3=pd.DataFrame(df.groupby(["Sub-Channels"])["Sales"].sum())
df3
```

```python
#4
df4=pd.DataFrame(df.groupby(["Month"])["Sales"].max())
df4=pd.DataFrame([df.sort_values(by=["Month","Sales"],ascending=False).iloc[0]['Month']])
df4
```

```python
#5
#dff=df.copy(deep=True)
df['Product Name clean'] = df['Product Name'].str.replace('\d+', '**')
df["Product Name clean"] = df["Product Name clean"].apply(lambda x: x.split("**")[0])
#df
df5=pd.DataFrame(df.groupby(["Product Name clean"]).sum()["Sales"]).sort_values(by=["Sales"],ascending=False)
#df5=pd.DataFrame([df5.sort_values(by=["Sales"],ascending=False).iloc[0]['Product Name clean']])
df5=pd.DataFrame(df5.iloc[0])
df5
```

```python
#6
df6=pd.DataFrame(df)
df6=pd.DataFrame([df["Product Name clean"].unique()])
df6.transpose()
```

```python
#7
with pd.ExcelWriter("fmcgassi.xlsx")as writer:
    df1.to_excel(writer, sheet_name='output1')
    df2.to_excel(writer, sheet_name='output2')
    df3.to_excel(writer, sheet_name='output3')
    df4.to_excel(writer, sheet_name='output4')
    df5.to_excel(writer, sheet_name='output5')
    df6.to_excel(writer, sheet_name='output6')
    
    
```
